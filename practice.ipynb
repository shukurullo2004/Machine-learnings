```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
labels = ['A', 'B', 'C', "D",'E' ]
values = [4, 8, 20,14,25]
hetches = ['/', '\\', '|', '-', '+', 'x', '.', '*']
plt.yticks(values)
bars = plt.bar(labels, values)
for bar in bars:
    bar.set_hatch(hetches.pop())
plt.figure(figsize=(6,4))
plt.show()
```
