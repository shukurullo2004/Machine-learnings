```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/Users/shukurullomeliboyev2004/Desktop/Machine learnings/matplotlib_tutorial-master/gas_prices.csv')

plt.figure(figsize=(10, 4))
plt.title("Gas prices over time  (in USD)", fontdict={'fontweight': 'bold', 'fontsize': 18})

# plt.plot(df['Year'], df.USA, 'b.-', label='USA')    
# plt.plot(df.Year, df.Australia,'r>--' ,label='Australia')
# plt.plot(df['Year'], df.Canada ,'g.-',label = 'Canada',)
# plt.plot(df['Year'], df['South Korea'], 'y.-', label='South Korea')

countries = ['Australia', 'USA', 'Canada', 'South Korea']

for country in countries:
    plt.plot(df.Year, df[country], marker='.', label=country) 

plt.xticks(df.Year[::3].tolist()+[2011])
plt.xlabel("Years")
plt.ylabel("US Dollars")
plt.legend()
plt.show()
plt.savefig('Gas_pricefigure.png', dpi=300)
```
